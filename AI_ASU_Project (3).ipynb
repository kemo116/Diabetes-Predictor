## Decision Tree

```python
from sklearn.tree import DecisionTreeClassifier #to import Decision Tree algorithm
from sklearn.metrics import classification_report # to import [f1_score , accuracy_score , precision_score , recall_score] report
```

```python
def tarin_using_entropy(X_train, y_train):
    
    # Decision tree with entropy
    clf_entropy = DecisionTreeClassifier(
            criterion = "entropy", random_state = 100)
    
    # Performing training
    clf_entropy.fit(X_train, y_train)
    
    return clf_entropy
```

```python
# Function to make predictions
def prediction(X_test, clf_object):
  
    # Predicton on test with Cross-Entropy
    y_pred = clf_object.predict(X_test)
    print("Predicted values:")
    print(y_pred)
    
    return y_pred
```

```python
# Function to calculate accuracy
def cal_accuracy(y_test, y_pred):
      
    print("Confusion Matrix: ",'\n',
        confusion_matrix(y_test, y_pred))
      
    print ("Accuracy : ",
    accuracy_score(y_test,y_pred)*100)
      
    print("Report : ",'\n',
    classification_report(y_test, y_pred))
```

```python
clf=tarin_using_entropy(X_train,y_train)
```

```python
y_pred= prediction(X_test, clf)
```

```python
cal_accuracy(y_test, y_pred)
```

```python
from sklearn.metrics import f1_score , accuracy_score , precision_score , recall_score
print("F1 score :",f1_score(y_pred,y_test))
print("accuracy score :",accuracy_score(y_pred,y_test))
print("precision : ",precision_score(y_pred,y_test))
print("recall :",recall_score(y_pred,y_test))
```

#Combining Models

```python
from statistics import mode
def combine(lr,nb,dtree,row):
  arr=list()
  arr.append(lr.predict(row))
  arr.append(nb.predict(row))
  arr.append(dtree(row))
  return mode(arr)[0]
```
